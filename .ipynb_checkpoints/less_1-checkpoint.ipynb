## Определение и характеристики


термин большие данные относится к наборам данных, размер которых превосходит возможности типичных баз данных (БД) по занесению, хранению, управлению и анализу информации.
Примеры:
- данные каналов Twitter, 
- данные посещаемости веб-страниц,
- данные мобильных приложений, 
- сетевой трафик, 
- данные датчиков

Принцип VVV означет налицие трёх признаков у больших данных:
- **Большой объем (Variable)** - В некоторые организации могут поступать десятки терабайт данных, в другие — сотни петабайт. Данные считаются большими, если их объем и сложность превышают возможности традиционных методов обработки и анализа, требуя использования специализированных инструментов и технологий для их эффективного управления и анализа.
- **Скорость (Volume)** - скорость приема данных и, возможно, действий на их основе. Обычно высокоскоростные потоки данных поступают прямо в оперативную память, а не записываются на диск
- **Разнообразие (Velocity)** - С появлением Big Data данные стали поступать в неструктурированном виде. Текст, аудио и видео.

## История больших данных

1) **1960-е — 1970-е**. В этот период началось развитие реляционных баз данных. Эдгар Кодд предложил реляционную модель данных, которая стала основой для современных систем управления базами данных (СУБД). Появились первые в мире ЦОД и реляционные базы данных.
2) **1980-е — 1990-е**. С развитием интернета начался экспоненциальный рост объема данных. Веб-сайты и онлайн-сервисы начали собирать огромные объемы данных о пользователях. Появление инструментов для извлечения, трансформации и загрузки данных (ETL), которые позволили эффективно управлять большими объемами данных.
3) **В 2005** году Дуг Каттинг и Майк Кэфри создали Hadoop, открытую платформу для распределенной обработки больших объемов данных. Hadoop стал основой для многих последующих технологий Больших данных. 
4) **2006 - 2010**. Появление NoSQL баз данных, таких как MongoDB и Cassandra, которые позволяли хранить и обрабатывать неструктурированные и полуструктурированные данные. Появление облачных платформ, таких как Amazon Web Services (AWS), Google Cloud Platform (GCP) и Microsoft Azure, которые предоставили мощные инструменты для хранения и обработки больших данных
5) **2008** Непосредственно термин «большие данные» появился в употреблении только в конце 2000-х. Он относится к числу немногих названий, имеющих вполне достоверную дату своего рождения — 3 сентября 2008 года, когда вышел специальный номер старейшего британского научного журнала Nature, посвященный поиску ответа на вопрос «Как могут повлиять на будущее науки технологии, открывающие возможности работы с большими объемами данных?». Специальный номер подытоживал предшествующие дискуссии о роли данных в науке вообще и в электронной науке (e-science) в частности
6) **2010**. С появлением Интернета вещей (IoT) все большее число устройств получает подключение к Интернету, что позволяет собирать данные о моделях действий пользователей и работе продуктов. А когда появились технологии машинного обучения, объем данных вырос еще больше.
7) **2010-е**. Большие данные стали основой для развития машинного обучения и искусственного интеллекта. Появление библиотек и фреймворков, таких как TensorFlow и PyTorch, позволило эффективно использовать большие данные для обучения моделей. Появление технологий потоковой обработки данных, таких как Apache Kafka и Apache Flink, которые позволили обрабатывать данные в реальном времени. Появление инструментов для визуализации данных, таких как Tableau и Power BI, которые позволили эффективно представлять и анализировать большие объемы данных.
8) **2020-е**. Появление инструментов для автоматизации процессов машинного обучения, таких как AutoML, которые позволяют упростить и ускорить процесс разработки моделей. Появление технологий edge computing, которые позволяют обрабатывать данные на периферийных устройствах, ближе к источнику данных, что уменьшает задержки и улучшает производительность. Усиление внимания к вопросам этики и приватности данных. Появление регуляций, таких как GDPR в Европе, которые накладывают строгие требования на обработку и хранение данных.

## Возникающие проблемы

- **Сбор данных**. Это могут быть открытые и внутренние источники.
- **Хранение данных**. Большие данные предсказуемо занимают много места. Хотя новые технологии хранения постоянно развиваются, объемы данных возрастают вдвое почти каждые два года. Организации до сих пор сталкиваются с проблемами роста объемов данных и их эффективного хранения.
- **Подготовка данных**. Недостаточно просто найти большое хранилище. Данные необходимо использовать, чтобы они приносили выгоду, и размер этой выгоды зависит от обработки данных. Чистые данные, то есть данные, актуальные для клиента и организованные для эффективного анализа, требуют тщательной обработки. Специалисты по изучению данных тратят от 50 до 80% рабочего времени на обработку и подготовку данных для использования.
- **постоянное развиете технологий**. технологии больших данных развиваются семимильными шагами. Несколько лет назад Apache Hadoop была самой популярной технологией для работы с большими данными. Платформа Apache Spark появилась в 2014 году. Сегодня оптимальным подходом является совместное использование этих двух платформ. Чтобы успевать за развитием больших данных, требуется прилагать большие усилия.
- **Интеграция данных**. Технология больших данных позволяет объединять данные из разрозненных источников и приложений. Традиционные механизмы интеграции, такие как средства для извлечения, преобразования и загрузки данных (ETL), не справляются с подобными задачами. Для анализа наборов данных размером в терабайт, а то и петабайт, нужны новые стратегии и технологии.
- **обновление**. 
- **Управление данными**. Большим данным требуется объемное хранилище. Решение для хранения может быть размещено в локальной или облачной среде или и там и там. Вы можете хранить данные в предпочтительном формате и применять желаемые требования к обработке (и необходимые механизмы обработки) к наборам данным по мере необходимости. Большинство организаций выбирают решение для хранения данных в зависимости от того, где они хранятся в настоящее время. Облачные хранилища пользуются растущей популярностью, так как поддерживают актуальные требования к вычислениям и позволяют задействовать ресурсы по мере надобности.
- **Анализ данных**
- **совместное использование**
- **передача**
- **визуализация**
